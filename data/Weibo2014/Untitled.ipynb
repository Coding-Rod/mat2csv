```python
def convert_mat3(file):
    data = scipy.io.loadmat(file)
    direction = file.split('.')[0]
    os.mkdir(direction)
    for i in range(data['data'].shape[-1]):
#     for i in range(2):
        df = pd.DataFrame(data['data'][:,:,i])
        df = df.set_index(0)
        df.to_csv(direction+'/data'+str(i)+'.csv')
    with open(direction+"/metadata.json","w") as f:
        f.write(json.dumps(dict(enumerate([int(list(x)[0]) for x in data['label']])), indent = 4))
    
```
